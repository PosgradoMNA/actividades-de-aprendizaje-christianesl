# Model Evaluation

Se recomienda separar los datos en entrenamiento y pruebas. 

la funcion train_test_split() nos ayuda a separar los datos. 

Cross validation 
Divides los datos en varias partes. En cada iteracion 1 parte se usa para entrenamiento, otra parte para validacion y la ultima parte para pruebas. 
En la siguiente iteracion se cambian de rol y asi sucesivamente. Cada parte es utilizada en general para entrenamiento, validacion & pruebas. 


sub entrenamiento & sobre entrenamiento

Cuando se entrena un modelo y la prediccion no se ajusta a los valores o lo hace de manera muy pobre es sub entrenamiento. 

Cuando el modelo esta sobre entrenado significa que la linea de prediccion esta casi en cada punto. Esto significa que el modelo se esta aprediendo los valores. 

En ambos casos se quiere evitar estos 2 tipos de entrenamientos porque no son buenos. 


#Ridge Regression

Un modelo que se utiliza para regresiones polinomiales. Evita los sobre entrenamientos. Utiliza una variable alfa. Dependiendo de los valores que se le asigne a alfa, el modelo tiene diferente comportamiento. 

Al utilizar este modelo, se tiene que probar diferentes valores para alfa para ver cual valor nos conviene. Dependiendo del valor puede dar un modelo sobre/sub entrenado tambien. 

# Grid Search

Los parametros que utilizan las funciones se llaman hyperparametros. Grid Search nos ayuda a encontrar los valors optimos.

```python
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))

    ax1 = sns.distplot(RedFunction, hist=False, color="r", label=RedName)
    ax2 = sns.distplot(BlueFunction, hist=False, color="b", label=BlueName, ax=ax1)

    plt.title(Title)
    plt.xlabel('Price (in dollars)')
    plt.ylabel('Proportion of Cars')

    plt.show()
    plt.close()
```

```python
def PollyPlot(xtrain, xtest, y_train, y_test, lr,poly_transform):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))
    xmax=max([xtrain.values.max(), xtest.values.max()])

    xmin=min([xtrain.values.min(), xtest.values.min()])

    x=np.arange(xmin, xmax, 0.1)


    plt.plot(xtrain, y_train, 'ro', label='Training Data')
    plt.plot(xtest, y_test, 'go', label='Test Data')
    plt.plot(x, lr.predict(poly_transform.fit_transform(x.reshape(-1, 1))), label='Predicted Function')
    plt.ylim([-10000, 60000])
    plt.ylabel('Price')
    plt.legend()
```

```python
Rsqu_test = []

order = [1, 2, 3, 4]
for n in order:
    pr = PolynomialFeatures(degree=n)
    
    x_train_pr = pr.fit_transform(x_train[['horsepower']])
    
    x_test_pr = pr.fit_transform(x_test[['horsepower']])    
    
    lr.fit(x_train_pr, y_train)
    
    Rsqu_test.append(lr.score(x_test_pr, y_test))

plt.plot(order, Rsqu_test)
plt.xlabel('order')
plt.ylabel('R^2')
plt.title('R^2 Using Test Data')
plt.text(3, 0.75, 'Maximum R^2 ')    
```

#Ridge Regression

```python
pr=PolynomialFeatures(degree=2)
x_train_pr=pr.fit_transform(x_train[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg','normalized-losses','symboling']])
x_test_pr=pr.fit_transform(x_test[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg','normalized-losses','symboling']])
```

```python
from sklearn.linear_model import Ridge
```

```python
RigeModel=Ridge(alpha=1)
RigeModel.fit(x_train_pr, y_train)
yhat = RigeModel.predict(x_test_pr)

print('predicted:', yhat[0:4])
print('test set :', y_test[0:4].values)
```

```python
from tqdm import tqdm

Rsqu_test = []
Rsqu_train = []
dummy1 = []
Alpha = 10 * np.array(range(0,1000))
pbar = tqdm(Alpha)

for alpha in pbar:
    RigeModel = Ridge(alpha=alpha) 
    RigeModel.fit(x_train_pr, y_train)
    test_score, train_score = RigeModel.score(x_test_pr, y_test), RigeModel.score(x_train_pr, y_train)
    
    pbar.set_postfix({"Test Score": test_score, "Train Score": train_score})

    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
```

```python
width = 12
height = 10
plt.figure(figsize=(width, height))

plt.plot(Alpha,Rsqu_test, label='validation data  ')
plt.plot(Alpha,Rsqu_train, 'r', label='training Data ')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.legend()
```

```python
RigeModel = Ridge(alpha=10) 
RigeModel.fit(x_train_pr, y_train)
RigeModel.score(x_test_pr, y_test)
```
